```python
import numpy as np 
import matplotlib.pyplot as plt
```

```python
S_0 = 100 #starting price
T = 1 #years
N= 252 #nummber of days to trade 
sigma = 0.2
mu = 0.1 #10percent
dt = T/N
```

```python
num_paths = 10
```

```python
all_paths = np.zeros((N, num_paths))

for j in range(num_paths):
    prices = np.zeros(N)
    prices[0] = S_0
    for t in range(1, N):
        Z = np.random.randn()
        prices[t] = prices[t-1] * np.exp((mu - 0.5*sigma**2)*dt 
                                         + sigma*np.sqrt(dt)*Z)
    all_paths[:, j] = prices
```

```python
plt.figure(figsize=(12,5))
plt.plot(all_paths)
plt.title("10 Simulated GBM Price Paths")
plt.xlabel("Time Step")
plt.ylabel("Price")
plt.show()
```

```python
final_prices = all_paths[-1, :]   # last row = final prices of all paths

plt.figure(figsize=(8,5))
plt.hist(final_prices, bins=30, edgecolor='black')
plt.title("Histogram of Final Prices")
plt.xlabel("Final Price")
plt.ylabel("Frequency")
plt.show()
```
